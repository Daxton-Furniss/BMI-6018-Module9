```python
#Question 1 
JFK_to_SLC = len(flights_data[(flights_data['origin'] == 'JFK') & (flights_data['dest'] == 'SLC')])
print("Number of flights from JFK to SLC:", JFK_to_SLC)
```

```python
#Question 2
Airlines_to_SLC = flights_data[flights_data['dest'] == 'SLC']['carrier'].nunique()
print("Number of Airlines to SLC':", Airlines_to_SLC)
```

```python
#Question 3
average_delay = flights_data[flights_data['dest'] == 'RDU']['arr_delay'].mean()
print("Average arrival delay for flights to RDU:", average_delay.round(2))
```

```python
#Question 4
sea_flights = flights_data[flights_data['dest'] == 'SEA']
total_sea_flights = len(sea_flights)
nyc_flights = sea_flights[(sea_flights['origin'] == 'LGA') | (sea_flights['origin'] == 'JFK')]
flights_from_lga_jfk = len(nyc_flights)
proportion = flights_from_lga_jfk / total_sea_flights
print(f"Proportion of flights from LGA and JFK to SEA: {proportion:.2%}")
```

```python
#Question 5
average_delay_per_date = flights_data.groupby('day')['dep_delay'].mean().reset_index()
date_and_largest_delay = average_delay_per_date[average_delay_per_date['dep_delay'] == average_delay_per_date['dep_delay'].max()]
date_and_largest_delay
```

```python
#Question 6
average_arrival_per_date = flights_data.groupby('day')['arr_delay'].mean().reset_index()
date_and_largest_arrival = average_arrival_per_date[average_arrival_per_date['arr_delay'] == average_arrival_per_date['arr_delay'].max()]
date_and_largest_arrival
```
